5. (11б). Створіть 2 дочірніх класи від Фігури: Трикутник та Коло, які мають атрибути, необхідні для розрахунку площин. Визначте метод get_area в кожному з них так, щоби вони розраховували площу в залежності від формули для кожного типу фігури. Створіть обʼєкт класу Трикутник та обʼєкт класу Коло. Виконайте операцію суми за допомогою оператора + між ними.

```python
import math

class Triangle(Figure):
    def __init__(self, a, h):
        self.a = a
        self.h = h

    def get_area(self):
        return 0.5 * self.a * self.h

class Circle(Figure):
    def __init__(self, r):
      self.r = r

    def get_area(self):
      return math.pi * pow(self.r, 2)

circle = Circle(6)
triangle = Triangle(5,10)
sum = circle + triangle
print(sum)
```

8. (16б). Створіть клас Library, який містить список об'єктів типу Book.
Реалізуйте методи для додавання книги, видалення книги та відображення списку книг.

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Книгу '{book.title}' додано до бібліотеки")

    def remove_book(self, title):
        for book in self.books:
            if book.title == title:
                self.books.remove(book)
                print(f"видалити {book.title}")
                return
        print(f"Книгу '{title}' не знайдено")

    def display_books(self):
        if not self.books:
            print("пусто")
        else:
            print("Книги:")
            for book in self.books:
                print(f"   {book.title}. Автор: {book.author}")

book1 = Book("Одіссея", "Гомер")
book2 = Book("Пригоди Гекльберрі Фінна", "Марк Твен")
book3 = Book("Гамлет", "Вільям Шекспір")

library = Library()
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
library.display_books()
library.remove_book("Гамлет")
library.display_books()
```
